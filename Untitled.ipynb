```python
import pandas as pd    
import numpy as np
import pylab
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import seaborn as sns
```

```python
fig, ax0 = plt.subplots(figsize=(10,6))
ICG = [0.95,4.48,22.6,32,515,168,126]
DH = [0.23,1.14,14.4,31.6,1181.4,321,804]
X=["J 2","J 3","J 7","J 18","J 25","J 19","J 23"]
plt.plot(X,ICG)
plt.plot(X,DH)
sns.set_style('ticks') # Arrière plan blanc
sns.despine() # Eliminer les bordures à droite et en haut
plt.xlabel("Instances",fontsize=14,fontweight='bold')
plt.ylabel("Temps en secondes",fontsize=14,fontweight='bold')
plt.tick_params(axis='x',labelsize=14)
plt.tick_params(axis='y',labelsize=14)
plt.title("Comparaison du temps de traitement entre ICG et DH",fontsize=16,fontweight='bold',fontstyle='italic',color='black')
plt.legend(labels=["ICG","DH"], loc='upper right',fontsize=14)
```
